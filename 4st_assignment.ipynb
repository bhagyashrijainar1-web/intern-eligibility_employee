```python
def performance_evaluation():
    total_rating = 0

    for i in range(1, 4):
        rating = int(input(f"Enter performance rating {i} (1-5): "))
        total_rating += rating

    average = total_rating / 3

    if average >= 4:
        result = "Excellent"
    elif average >= 3:
        result = "Good"
    else:
        result = "Needs Improvement"

    return average, result
```
